```
import numpy as np
```

```
def f1(x,y):
  return (x-3)**2+2*y
def f2(x,y):
  return np.sin(x-3)**2+2*np.cos(y-1)
```

```
#Optimize over the range 0-10,0-5
#find minimum value
```

```
def hill_climb(f,step_size):
  x_start=np.random.rand()*10
  y_start=np.random.rand()*5
  for i in range(1000):

    test_xh,test_xl=np.minimum(10.0,x_start+step_size),np.maximum(0.0,x_start-step_size)
    test_yh,test_yl=np.minimum(5.0,y_start+step_size),np.maximum(0.0,y_start-step_size)
    input_vals=[(x_start,y_start),(test_xh,test_yh),(test_xh,test_yl),(test_xl,test_yh),(test_xl,test_yl)]
    vals=[f(x_start,y_start),f(test_xh,test_yh),f(test_xh,test_yl),f(test_xl,test_yh),f(test_xl,test_yl)]
    m1=np.argmin(vals)
    x_start,y_start=input_vals[m1]
  return x_start,y_start  
    
```

```
hill_climb(f2,.1)
```
