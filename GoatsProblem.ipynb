```
import numpy as np
class Queue:
  def __init__(self):
     self.items = []

  def isEmpty(self):
     return self.items == []

  def push(self, item):
     self.items.append(item)

  def pop(self):
    val=self.items[0]
    self.items=self.items[1:]
    return val
      
  def peek(self):
     return self.items[0]

  def size(self):
     return len(self.items)
```

```
class goat_problem():
  def __init__(self):
    self.state=[1,1,1,1]
    self.names=["farmer","goat","wolf","cabbage"]
    self.visited_states={}

  def legal_moves(self,curr_state):
    possible_next_states=[]
    farmer_next_state=1-curr_state[0]
    
    trial_state=curr_state.copy()
    trial_state[0]=farmer_next_state
    possible_next_states.append(trial_state)
    a=1
    for animal in curr_state[1:]:
      trial_state=curr_state.copy()
      trial_state[0]=farmer_next_state
      if(animal==curr_state[0]):
        trial_state[a]=1-curr_state[0]
        possible_next_states.append(trial_state)
      a=a+1
    legal_states=[]
    for state in possible_next_states:
      if(state[1]==state[2] and state[0]!=state[1]):
        continue
      if(state[1]==state[3] and state[0]!=state[1]):
        continue
      legal_states.append(state)

    return legal_states
  
  def solve_problem(self):
    frontier=Queue()
    frontier.push(self.state)
    while(frontier!=[]):
      if(frontier.items==[]):
        break
      current_node=frontier.pop()
      print(current_node)
      if(np.sum(current_node)==0):
        print("Solved")
        break
      for child in self.legal_moves(current_node):
        if str(child) in self.visited_states:
          continue
        self.visited_states[str(child)]=1
        frontier.push(child)
```
