```
#In this assignment you will code an agent to play iterated prisonners dilemma
# The payoffs are as follows
#     C     D
# C  2,2   0,3
# D  3,0   1,1


#Confessing when the other person confesses gives both a reward of 2,2 
#confessing when the other person doesnt gives you a reward of 0 and them a reward of 3
#Defecting when the other person confessing gives you a reward of 3 and them a reward of 0
#Both Defecting gives each a reward of 1
#Your goal is to maximize reward over a series of games
import numpy as np
#1 is defect
#0 is confess
```

```
class randomPlayer():
  def __init__(self):
    self.total_reward=0
  def choose_move(self):
     return np.random.randint(0,2)
  def update_internally(self,result): 
    self.total_reward+=result
    pass
  def reset(self):
    pass
  
class greedyPlayer():
  def __init__(self):
    self.total_reward=0
  def choose_move(self):
     return 1   
  def update_internally(self,result): 
    self.total_reward+=result
    pass
  def reset(self):
    pass  
class friendlyPlayer():
  def __init__(self):
    self.total_reward=0
  def choose_move(self):
     return 0
  def update_internally(self,result): 
    self.total_reward+=result
    pass
  def reset(self):
    pass  
class elephantPlayer():
  def __init__(self):
    self.total_reward=0
    self.betrayed=0
  def choose_move(self):
    if self.betrayed:
      return 1
    return 0
  def update_internally(self,result): 
    self.total_reward+=result 
    if result==0 :
      self.betrayed=1
  def reset(self):
    self.betrayed=0    

class copyPlayer():
  def __init__(self):
    self.total_reward=0
    self.last_played=0
  def choose_move(self):
    return self.last_played  
  def update_internally(self,result): 
    self.total_reward+=result 
    if result==0 :
      self.last_played=1
      return
    if result==1 :
      self.last_played=1
      return
    else:
      self.last_played=0
      return
  def reset(self):
    self.last_played=0    
    
def your_player():
  def __init__(self):
    self.total_reward=0
  def choose_move(self):
    pass
  def update_internally(self,result): 
    self.total_reward+=result 
  def reset(self):
    pass
```

```
#Try to get the maximum score by coding your player. You cannot use the loop to figure out what kind of player you are playing. You can only use the result to update your internal state. 
#The best agents will play against each other
```

```
player1=randomPlayer()
player2=greedyPlayer()
player3=friendlyPlayer()
player4=elephantPlayer()
player5=copyPlayer()
players=[player1,player2,player3,player4,player5]
my_player=copyPlayer()

for player in players:
  player.reset()
  my_player.reset()
  for _ in range(10):
    move1=my_player.choose_move()
    move2=player.choose_move()
    print(move1,move2)
    if(move1==0 and move2==0):
      result=2
      my_player.update_internally(result)
      player.update_internally(result)
      continue
    if(move1==1 and move2==1):
      result=1
      my_player.update_internally(result)
      player.update_internally(result)
      continue
    if(move1==1 and move2==0):
      my_player.update_internally(3)
      player.update_internally(0)
      continue
    if(move1==0 and move2==1):
      my_player.update_internally(0)
      player.update_internally(3)
      continue
  print(my_player.total_reward)
```
